```python
# URL of the men's singles stats page
url = 'http://www.espn.com/tennis/stats'

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content of the page using BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Find all players in the HTML content
players = soup.find_all('tr', attrs={'class': re.compile('row player-10-')})

# Define the columns of the DataFrame
columns = ['Player', 'Matches', 'Sets', 'Games', 'Aces', 'Double Faults', '1st Serve %', '1st Serve Points Won %', '2nd Serve Points Won %', 'Break Points Saved %', 'Service Games Won %', 'Return Games Won %', 'Break Points Converted %']

# Create an empty DataFrame to store the stats
final_df = pd.DataFrame(columns=columns)

# Extract the stats for each player and add them to the DataFrame
for player in players:
    stats = [stat.get_text() for stat in player.find_all('td')]
    temp_df = pd.DataFrame(stats).transpose()
    temp_df.columns = columns
    final_df = pd.concat([final_df, temp_df], ignore_index=True)

# Print the DataFrame
print(final_df)
```

```python
# URL of the women's singles stats page
url = 'http://www.espn.com/tennis/stats/_/type/women'

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content of the page using BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Find all players in the HTML content
players = soup.find_all('tr', attrs={'class': re.compile('row player-10-')})

# Define the columns of the DataFrame
columns = ['Player', 'Matches', 'Sets', 'Games', 'Aces', 'Double Faults', '1st Serve %', '1st Serve Points Won %', '2nd Serve Points Won %', 'Break Points Saved %', 'Service Games Won %', 'Return Games Won %', 'Break Points Converted %']

# Create an empty DataFrame to store the stats
final_df = pd.DataFrame(columns=columns)

# Extract the stats for each player and add them to the DataFrame
for player in players:
    stats = [stat.get_text() for stat in player.find_all('td')]
    temp_df = pd.DataFrame(stats).transpose()
    temp_df.columns = columns
    final_df = pd.concat([final_df, temp_df], ignore_index=True)

# Print the DataFrame
print(final_df)
```

### Win/Loss Record on the ATP Tour

```python
# URL of the ATP win-loss records page
url = 'https://www.atptour.com/en/stats/win-loss-index'

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content of the page using BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Find all players in the HTML content
players = soup.find_all('tr', attrs={'class': re.compile('stats-table-row')})

# Define the columns of the DataFrame
columns = ['Rank', 'Player', 'Win-Loss Record']

# Create an empty DataFrame to store the stats
final_df = pd.DataFrame(columns=columns)

# Extract the win-loss records for each player and add them to the DataFrame
for player in players:
    rank = player.find('td', attrs={'class': 'rank-cell'}).get_text()
    name = player.find('td', attrs={'class': 'player-cell'}).get_text()
    record = player.find('td', attrs={'class': 'win-loss-cell'}).get_text()
    temp_df = pd.DataFrame([[rank, name, record]], columns=columns)
    final_df = pd.concat([final_df, temp_df], ignore_index=True)

# Print the DataFrame
print(final_df)
```

### Win/Loss Record on the WTA Tour

```python
# URL of the WTA win-loss records page
url = 'https://www.wtatennis.com/stats'

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content of the page using BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Find all players in the HTML content
players = soup.find_all('tr', attrs={'class': re.compile('stats-table-row')})

# Define the columns of the DataFrame
columns = ['Rank', 'Player', 'Win-Loss Record']

# Create an empty DataFrame to store the stats
final_df = pd.DataFrame(columns=columns)

# Extract the win-loss records for each player and add them to the DataFrame
for player in players:
    rank = player.find('td', attrs={'class': 'rank-cell'}).get_text()
    name = player.find('td', attrs={'class': 'player-cell'}).get_text()
    record = player.find('td', attrs={'class': 'win-loss-cell'}).get_text()
    temp_df = pd.DataFrame([[rank, name, record]], columns=columns)
    final_df = pd.concat([final_df, temp_df], ignore_index=True)

# Print the DataFrame
print(final_df.to_csv)
```
